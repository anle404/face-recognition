```python
import joblib
import os
import numpy as np
from PIL import Image
import time
import datetime
import gc
from tqdm import tqdm
```

```python
IMAGE_SIZE = 64
IMAGE_DIMENSIONS = (IMAGE_SIZE, IMAGE_SIZE)
IMAGE_SHAPE = (IMAGE_SIZE, IMAGE_SIZE, 3)

def load_images(folder_path, target_size=IMAGE_DIMENSIONS):
    images = []
    labels = []

    for image_dir in os.listdir(folder_path):
        current_dir = os.path.join(folder_path, image_dir)
        for filename in os.listdir(current_dir):
            img_path = os.path.join(current_dir, filename)
            try:
                img = Image.open(img_path).convert("RGB")
                img = img.resize(target_size)
                img = np.array(img)
                images.append(img)
                labels.append(image_dir)
            except Exception as e:
                print(f"Error loading {img_path}: {e}")

    return np.array(images), np.array(labels)

def load_verification(folder_path, target_size=IMAGE_DIMENSIONS):
    images = []
    labels = []

    for filename in os.listdir(folder_path):
        img_path = os.path.join(folder_path, filename)
        try:
            img = Image.open(img_path).convert("RGB")
            img = img.resize(target_size)
            img = np.array(img)
            images.append(img)
            dir_name = os.path.basename(os.path.dirname(img_path))
            file_name = os.path.basename(img_path)
            labels.append(f"./{dir_name}/{file_name}")
        except Exception as e:
            print(f"Error loading {img_path}: {e}")

    return np.array(images), np.array(labels)
```

```python
TRAINING_DIR = "./Assignment2/data/classification_data/train_data"
VALIDATION_DIR = './Assignment2/data/classification_data/val_data'
TESTING_DIR = './Assignment2/data/classification_data/test_data'
VERIFICATION_DIR = './Assignment2/data/verification_data'
OUTPUT_DIR = './Assignment2/pickles'
```

```python
print("Loading training...")
training_images, training_labels = load_images(TRAINING_DIR)
```

```python
print(training_images.shape)
print(training_labels.shape)
```

```python
print("Loading validation...")
validation_images, validation_labels = load_images(VALIDATION_DIR)
```

```python
print(validation_images.shape)
print(validation_labels.shape)
```

```python
print("Loading testing...")
testing_images, testing_labels = load_images(TESTING_DIR)
```

```python
print(testing_images.shape)
print(testing_labels.shape)
```

```python
print("Loading verification...")
verification_images, verification_labels = load_verification(VERIFICATION_DIR)
```

```python
print(verification_images.shape)
print(verification_labels.shape)
```

```python
data_to_pickle = [
    ("training_images.pkl", training_images),
    ("training_labels.pkl", training_labels),
    ("validation_images.pkl", validation_images),
    ("validation_labels.pkl", validation_labels),
    ("testing_images.pkl", testing_images),
    ("testing_labels.pkl", testing_labels),
    ("verification_images.pkl", verification_images),
    ("verification_labels.pkl", verification_labels)
]

for filename, data in data_to_pickle:
    print(f"Creating {filename}")
    path = os.path.join(OUTPUT_DIR, filename)
    if data is not None:
      print('data is not None')
      print(type(data))
      with open(path, 'wb') as f:
        joblib.dump(data, f, compress=0)
```
